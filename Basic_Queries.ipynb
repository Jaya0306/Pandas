```python
import pandas as pd
```

```python
# Creating the DataFrame

# Expanding the DataFrame
data = {
    'id': list(range(1, 21)),  # Employee IDs from 1 to 20
    'name': [
        'Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Henry', 
        'Ivy', 'Jack', 'Kevin', 'Laura', 'Mike', 'Nancy', 'Oliver', 'Paul', 
        'Quinn', 'Rachel', 'Steve', 'Tina'
    ],
    'department': [
        'HR', 'IT', 'Finance', 'IT', 'HR', 'IT', 'Finance', 'IT', 
        'Marketing', 'Sales', 'Finance', 'Sales', 'IT', 'HR', 'Marketing', 'IT', 
        'Finance', 'Sales', 'Marketing', 'HR'
    ],
    'salary': [
        60000, 85000, 75000, 92000, 54000, 48000, 97000, 110000, 
        68000, 73000, 89000, 72000, 95000, 51000, 67000, 88000, 
        93000, 76000, 70000, 59000
    ],
    'joining_date': pd.to_datetime([
        '2021-06-15', '2020-09-23', '2018-02-14', '2019-11-01', '2022-05-10', 
        '2021-07-20', '2017-08-25', '2016-12-05', '2020-04-17', '2019-03-08', 
        '2015-06-25', '2021-09-30', '2018-11-20', '2019-07-15', '2017-02-28', 
        '2016-10-05', '2020-12-11', '2018-05-22', '2019-08-13', '2022-01-19'
    ]),
    'city': [
        'New York', 'San Francisco', 'Chicago', 'Austin', 'New York', 'San Francisco', 'Chicago', 'Austin', 
        'Seattle', 'Boston', 'Chicago', 'Boston', 'Austin', 'New York', 'Seattle', 'San Francisco', 
        'Chicago', 'Boston', 'Seattle', 'New York'
    ]
}
```

```python
df = pd.DataFrame(data)
```

#### Select All Data

```python
df #SELECT * FROM employees;
```

#### Select Specific Columns (Only Name & Salary)

```python
df[['name','salary']] #SELECT name, salary FROM employees;
```

#### Filter Employees with Salary Greater than 70,000

```python
df[df['salary'] > 70000] #SELECT * FROM employees WHERE salary > 70000;
```

####  Filter Employees from New York or Austin

```python
df[df['city'].isin(['New York', 'Austin'])] #SELECT * FROM employees WHERE city IN ('New York', 'Austin');
```

####  Filter Employees Who Joined After 2020-01-01

```python
df[df['joining_date'] > '2020-01-01'] #SELECT * FROM employees WHERE joining_date > '2020-01-01';
```

#### Find employees with salaries between 50,000 and 90,000.

```python
df[(df['salary'] >= 50000) & (df['salary'] <= 90000)] #SELECT * FROM employees WHERE salary BETWEEN 50000 AND 90000;
```

#### Find employees whose names start with 'A'.

```python
df[df['name'].str.startswith('A')] #SELECT * FROM employees WHERE name LIKE 'A%';
```

#### Find employees whose names end with 'e'.

```python
df[df['name'].str.endswith('e')] #SELECT * FROM employees WHERE name LIKE '%e';
```

#### Find employees whose names contain 'ar'.

```python
df[df['name'].str.contains('ar')] #SELECT * FROM employees WHERE name LIKE '%ar%';
```

#### Find employees who are not in the IT department.

```python
df[df['department'] != 'IT'] #SELECT * FROM employees WHERE department <> 'IT';
```

#### Find employees whose salary is NULL (missing values).

```python
df[df['salary'].isnull()] #SELECT * FROM employees WHERE salary IS NULL;
```

```python
df[df['salary'].notnull()] #SELECT * FROM employees WHERE salary IS NOT NULL;
```

#### Find employees in the IT department AND with a salary greater than 80,000.

```python
df[(df['department'] == 'IT') & (df['salary'] > 80000)]  #SELECT * FROM employees WHERE department = 'IT' AND salary > 80000;
```

#### Find employees who are either in HR or Finance.

```python
df[(df['department'] == 'HR') | (df['department'] == 'Finance')] #SELECT * FROM employees WHERE department = 'HR' OR department = 'Finance';
```

#### Get a list of all unique departments.

```python
df['department'].unique() #SELECT DISTINCT department FROM employees;
```

#### Get all employees sorted by salary in descending order.

```python
df.sort_values(by='salary',ascending=False) #SELECT * FROM employees ORDER BY salary DESC;
```

#### Get the top 3 highest-paid employees.

```python
df.nlargest(3,'salary')
```
